# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

2、再筛选出UnitPrice为负数的观察值

评估结论5-2：UnitPrice为负数的观察值对应的Description都是'Adjust bad debt'，说明不是实际商品交易数据，后续清洗过程中应删除

## 清理数据

首先根据上面评估部分得到的结论，总结我们需要进行的数据清理任务：
* 修改'CustomerID'变量的数据类型为str
* 修改'InvoiceDate'变量的数据类型为日期时间
* 将'Description'变量缺失的观察值删除
* 将'Country'变量值'USA'替换为'United States'
* 将'Country'变量值'UK','U.K.'替换为'United Kingdom'
* 将'Quantity'变量值为负数的观察值删除
* 将'UnitPrice'变量值为负数的观察值删除

```python
cleaned_data = original_data
```

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].astype('string')
# 旧版 pandas 里字符串列是 object，新版可用 'string' 省内存且支持 <NA>
print(cleaned_data['CustomerID'].dtype)
cleaned_data['CustomerID']
```

```python
# 观察到，即使数据类型转换正确之后，CustomerID还是以.0结尾，下面将其删除
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].str.slice(0,5)
# slice语法同切片，右边不保留
cleaned_data['CustomerID']
# cleaned_data.head()
```

```python
cleaned_data['InvoiceDate'] = pd.to_datetime(cleaned_data['InvoiceDate'])
cleaned_data['InvoiceDate'] 
```

```python
cleaned_data.dropna(subset='Description',inplace=True)
# 验证是否删除成功
cleaned_data['Description'].isnull().sum()
```

```python
cleaned_data['Country'] = cleaned_data['Country'].replace({
    'USA': 'United States',
    'UK': 'United Kingdom',
    'U.K.': 'United Kingdom'
})
# 验证是否替换成功
cleaned_data['Country'].value_counts()
# method2:cleaned_data[cleaned_data['Country']=='USA']
```

```python
cleaned_data = cleaned_data[(cleaned_data['Quantity']>=0)\
                            & (cleaned_data['UnitPrice']>=0)]
# 验证是否删除成功
len(cleaned_data[cleaned_data['Quantity']<0])
cleaned_data.head()
```

## 保存清理后的数据

因为观察值的索引就是普通的默认位置索引，可以不保存

```python
cleaned_data.to_csv("cleaned_e_commerce.csv",index = False)
```
